```python
import json
fd = open("record.json",'r')
r = fd.read()
fd.close()

record = json.loads(r)
```

```python
prod_id = str(input("Enter product id:"))
name = str(input("Enter name:"))
qn = int(input("Enter scoop:"))
cone = int(input("Enter price per one scoop in cone:"))
cup=int(input("Enter price per one sccop in cup:"))
exp = input("Enter expiry data(yyyy-mm-dd):")
cat= input("Enter category:")
vID=int(input("Enter vendorID:"))

for i in record:
    if(prod_id==i):
        qn=qn+record[prod_id]['quantity']
record[prod_id] = {'name': name,'quantity':qn,'cone':cone,'cup':cup,'expiry':exp,'category':cat,'vendorID':vID}
js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```
