# Billing section

```python
import json

fd = open("record.json",'r')
r = fd.read()
fd.close()

records = json.loads(r)
```

```python
fs = open("sale.json",'r')
s = fs.read()
fs.close()
update = json.loads(s)
txn=0
for i in update.keys():
    if(int(i)>int(txn)):
        txn=i
txn=int(txn)+1
```

```python
import time
pr=0
t=time.ctime()
print("****************ICE CREAM PARLOUR*******************\n")
while(True):
    ui_prod  = str(input("Enter the product_Id: "))
    ui_quant = int(input("Enter the number of scoops "))
    ui_price = input("Enter cone or cup:")
    if(ui_quant>records[ui_prod]['quantity']):
        ui_quant=records[ui_prod]['quantity']
        print("We regret to inform you that only "+str(ui_quant)+" number of scoop(s) are available")
        dec=input("Do you want to continue with this order?(y/n)")
        if(dec!='y'):
            break
    print("Product: ", records[ui_prod]['name'])
    if(ui_price=="cone"):
        price=records[ui_prod]['cone']
    else:
        price=records[ui_prod]['cup']
    print("Price per scoop: ", price)
    print("Category:",records[ui_prod]['category'])
    print("Expiry:",records[ui_prod]['expiry'])
    print("Amount:",price * ui_quant)
    ch=input("Do you want to purchase more?(y/n)")
    pr=pr+(price * ui_quant)
    
    records[ui_prod]['quantity'] = records[ui_prod]['quantity'] - ui_quant
    
    fs = open("sale.json",'r')
    s = fs.read()
    fs.close()
    update = json.loads(s)
    update[txn] = {'name':records[ui_prod]['name'],'productID':ui_prod, 'quantity' : ui_quant, 'amount': price * ui_quant,'ui_time':t}
    s = json.dumps(update)
    fs = open("sale.json",'w')
    fs.write(s)
    fs.close()
    txn+=1
    
    
    if(ch=='y'):
        print("\n\n")
    else:
        break
print("-------------------------------------------------------")
print("Billing Amount: ", pr)
print("-------------------------------------------------------")
print("Hope this scoop of ice cream makes your day great!\n                      Visit again")

print(t)

```
